```python
import pandas as pd
import numpy as np

# Loading the files to Jupiter notebook
Schools = "schools_complete.csv"
Students = "students_complete.csv"

# Reading students, and school files
District_School = pd.read_csv(Schools)
District_Students = pd.read_csv(Students)

# Merging all data into one file
District_complete = pd.merge(District_School, District_Students, how ="left", on = ["school_name" , "school_name"])

# Printing the 5 first elements of the merged table
District_complete.head()
```

```python
# Calculating values for the District Summary: Dissummary_df
Total_schools = len(District_complete["school_name"].unique())
Total_students = "{:,}".format(len(District_complete["student_name"]))
Total_Budget = "${:,}.00".format(round(District_complete["budget"].unique().sum(),2))
Average_MathScore = round(District_complete["math_score"].mean(),2)
Average_ReadingScore = round(District_complete["reading_score"].mean(),2)
Passing_math = round((((District_complete["math_score"] >= 70).sum())/len(District_complete["math_score"]))*100,2)
Passing_reading = round((((District_complete["reading_score"] >= 70).sum())/len(District_complete["reading_score"]))*100,2)
Overall_Passrate = round((Passing_math + Passing_reading)/2,2)
```

```python
# Dictionary creation for the DataFrame District Summary: Dissummary_df
Dissummary_df = pd.DataFrame({"Total Schools": [Total_schools],
                              "Total Students": [Total_students],
                              "Total Budget $": str(Total_Budget),
                              "Average Math Score": [Average_MathScore],
                              "Average Reading Score": [Average_ReadingScore],
                              "% Passing Math": str(Passing_math)+ "%",
                              "% Passing Reading": str(Passing_reading) + "%",
                              "Overall Passing Rate": str(Overall_Passrate)+ "%"})
#Printing the District Summary as DataFrame
Dissummary_df
```

```python
# From the Merged table data, reducing all entries for student information to just
# schools names in order to create the Schoo Summary Table as School_S
School_S= District_complete[['school_name', 'type']].drop_duplicates(keep='last')
# Changing the type column name to School Type
School_S.rename(columns={'type':'School Type'}, inplace=True)
# Assigning the index from the entry numbers to the column 'school_name'
School_S.set_index(('school_name'), inplace=True)
# Adding columns for future information with zero value:
School_S['Total Students']=0
School_S['Total School Budget'] = 0
School_S['Per Student Budget'] = 0
School_S['Average Math Score'] = 0
School_S['Average Reading Score'] = 0
```

```python
# Calculating the average value by school groups, obtaining with this the number of students
# by school, their budget per school, the average passing score for reading and math.
School_mean = District_complete.groupby(["school_name"]).mean()
School_S['Total Students'] = School_mean['size'].astype(float).map("{:,.0f}".format)
School_S['Total School Budget'] = School_mean['budget'].astype(float).map("${:,.2f}".format)
School_S['Per Student Budget'] = (School_mean['budget'].astype(float)/School_mean['size'].astype(float)).map("${:,.2f}".format)
School_S['Average Math Score'] = School_mean['math_score'].astype(float).map("{:.3f}".format)
School_S['Average Reading Score'] = School_mean['reading_score'].astype(float).map("{:.3f}".format)
```

```python
# Extracting information per High School, to calculate the Percentage of students passing
# Math, Reading and with those columns calculate the average for the Percentage Overall 
# Passing Rate.
Readingscore_df= District_complete.groupby('school_name')
g = Readingscore_df[['school_name', 'reading_score', 'math_score']]
Bailey= g.get_group('Bailey High School')
Cabrera= g.get_group('Cabrera High School')
Figueroa= g.get_group('Figueroa High School')
Ford= g.get_group('Ford High School')
Griffin= g.get_group('Griffin High School')
Hernandez= g.get_group('Hernandez High School')
Holden= g.get_group('Holden High School')
Huang= g.get_group('Huang High School')
Johnson= g.get_group('Johnson High School')
Pena= g.get_group('Pena High School')
Rodriguez= g.get_group('Rodriguez High School')
Shelton= g.get_group('Shelton High School')
Thomas= g.get_group('Thomas High School')
Wilson= g.get_group('Wilson High School')
Wright= g.get_group('Wright High School')
# Calculating the Percent Passing Math
Pmath_Bailey = round((((Bailey["math_score"] >= 70).sum())/len(Bailey["math_score"]))*100,3)
Pmath_Cabrera = round((((Cabrera["math_score"] >= 70).sum())/len(Cabrera["math_score"]))*100,3)
Pmath_Figueroa = round((((Figueroa["math_score"] >= 70).sum())/len(Figueroa["math_score"]))*100,3)
Pmath_Ford = round((((Ford["math_score"] >= 70).sum())/len(Ford["math_score"]))*100,3)
Pmath_Griffin = round((((Griffin["math_score"] >= 70).sum())/len(Griffin["math_score"]))*100,3)
Pmath_Hernandez = round((((Hernandez["math_score"] >= 70).sum())/len(Hernandez["math_score"]))*100,3)
Pmath_Holden = round((((Holden["math_score"] >= 70).sum())/len(Holden["math_score"]))*100,3)
Pmath_Huang = round((((Huang["math_score"] >= 70).sum())/len(Huang["math_score"]))*100,3)
Pmath_Johnson = round((((Johnson["math_score"] >= 70).sum())/len(Johnson["math_score"]))*100,3)
Pmath_Pena = round((((Pena["math_score"] >= 70).sum())/len(Pena["math_score"]))*100,3)
Pmath_Rodriguez = round((((Rodriguez["math_score"] >= 70).sum())/len(Rodriguez["math_score"]))*100,3)
Pmath_Shelton = round((((Shelton["math_score"] >= 70).sum())/len(Shelton["math_score"]))*100,3)
Pmath_Thomas = round((((Thomas["math_score"] >= 70).sum())/len(Thomas["math_score"]))*100,3)
Pmath_Wilson = round((((Wilson["math_score"] >= 70).sum())/len(Wilson["math_score"]))*100,3)
Pmath_Wright = round((((Wright["math_score"] >= 70).sum())/len(Wright["math_score"]))*100,3)
# Passing Percent Math list 
Passing_Math = [Pmath_Bailey, Pmath_Cabrera, Pmath_Figueroa, Pmath_Ford, Pmath_Griffin, Pmath_Hernandez, 
               Pmath_Holden, Pmath_Huang, Pmath_Johnson, Pmath_Pena, Pmath_Rodriguez, Pmath_Shelton, 
               Pmath_Thomas, Pmath_Wilson, Pmath_Wright]
# Calculating the Percent Passing Reading 
Preading_Bailey = round((((Bailey["reading_score"] >= 70).sum())/len(Bailey["reading_score"]))*100,3)
Preading_Cabrera = round((((Cabrera["reading_score"] >= 70).sum())/len(Cabrera["reading_score"]))*100,3)
Preading_Figueroa = round((((Figueroa["reading_score"] >= 70).sum())/len(Figueroa["reading_score"]))*100,3)
Preading_Ford = round((((Ford["reading_score"] >= 70).sum())/len(Ford["reading_score"]))*100,3)
Preading_Griffin = round((((Griffin["reading_score"] >= 70).sum())/len(Griffin["reading_score"]))*100,3)
Preading_Hernandez = round((((Hernandez["reading_score"] >= 70).sum())/len(Hernandez["reading_score"]))*100,3)
Preading_Holden = round((((Holden["reading_score"] >= 70).sum())/len(Holden["reading_score"]))*100,3)
Preading_Huang = round((((Huang["reading_score"] >= 70).sum())/len(Huang["reading_score"]))*100,3)
Preading_Johnson = round((((Johnson["reading_score"] >= 70).sum())/len(Johnson["reading_score"]))*100,3)
Preading_Pena = round((((Pena["reading_score"] >= 70).sum())/len(Pena["reading_score"]))*100,3)
Preading_Rodriguez = round((((Rodriguez["reading_score"] >= 70).sum())/len(Rodriguez["reading_score"]))*100,3)
Preading_Shelton = round((((Shelton["reading_score"] >= 70).sum())/len(Shelton["reading_score"]))*100,3)
Preading_Thomas = round((((Thomas["reading_score"] >= 70).sum())/len(Thomas["reading_score"]))*100,3)
Preading_Wilson = round((((Wilson["reading_score"] >= 70).sum())/len(Wilson["reading_score"]))*100,3)
Preading_Wright = round((((Wright["reading_score"] >= 70).sum())/len(Wright["reading_score"]))*100,3)
# Passing Percent Reading List 
Passing_Reading = [Preading_Bailey, Preading_Cabrera, Preading_Figueroa, Preading_Ford, Preading_Griffin,
                   Preading_Hernandez, Preading_Holden, Preading_Huang, Preading_Johnson, Preading_Pena, Preading_Rodriguez,
                   Preading_Shelton, Preading_Thomas, Preading_Wilson, Preading_Wright]
```

```python
# Our bins for the data spending based on the max and minimum expended per school, the allocation of the money as we 
# is not by school size, but for student demographics, some more need population require special services and those
# citizens requires some extra founding in some cases, like homeless or not english speaking population. Here,
# we are analizing the grade performance per student budget expended. Here we can see that those schools with less
# budget reflects higher grades than those with larger budget, because those are the schools with struggling kids,
#even with all the support invested they are as high as those with a more stable family support.
bins= [0, 500, 550, 600, 650, 700]
group_name= ["Less than 500", "between 500 and 550", "Between 550 and 600", "Between 600 and 650", "More than 650"]
School_spending["School spenditure frame"]= pd.cut(School_spending["Per Student Budget"], bins, labels=group_name)
```

```python
# Grade scores by school sizes from 427 as a minimum and with a maximun of 4976 students, the grade score are higher
# for schools with smaller populations.
bins= [0, 700, 1700, 2700, 3700, 4700]
group_name= ["Less than 700", "between 700 and 1700", "Between 1700 and 2700", "Between 2700 and 3700", "More than 3700"]
School_spending["School Size Intervals"]= pd.cut(School_spending["Total Students"], bins, labels=group_name)
```
